```python
import re
import unicodedata
```

```python
def process_text(text , lang):
    """ docstring """
    text = str(text)
    
    if lang == 'en':
        text = text.lower()
     
    # normalize unicode (NFKC = compatibility and Composition)
    text = unicodedata.normalize('NFKC',text) 
    
    # remove spaces
    text = re.sub(r"\s+", " ", text)
    
    # Remove weird control characters
    text = re.sub(r"[\u0000-\u001F\u007F-\u009F]", "", text)
    
    
    return text.strip()
    
        
```

```python
df['english'] = df['english'].apply(lambda x:process_text(x,'en'))
df['hindi'] = df['hindi'].apply(lambda x:process_text(x,'hi'))
```

# Dataset class

```python
import torch
```

```python
from torch.utils.data import Dataset

class CustomDataset(Dataset):
    """ docstring """
    def __init__(self,df):
        self.source = df['en_ids'].to_list() # explain why to_list() is used
        self.target = df['hi_ids'].to_list()
     
    def __len__(self):
        return len(self.source)  
    
    def __getitem__(self, index):
        return (
            torch.tensor(self.source[index]),
            torch.tensor(self.target[index])
            )  
        
```

### Collate function for padding

```python
import torch.nn.utils.rnn as rnn

def collate_fn(batch):
    source_batch, target_batch = zip(*batch)
    
    source_batch = rnn.pad_sequence(source_batch, batch_first= True, padding_value=0)
    target_batch = rnn.pad_sequence(target_batch, batch_first= True, padding_value=0)
    
    return source_batch,target_batch
```

## Tranformer Model

```python
import torch.nn as nn

class Custom_Transformer(nn.Module):
    
    def __init__(self, vocab_size, d_model, nhead, num_layers):
         super().__init__()
         
         self.embedding = nn.Embedding(vocab_size,d_model)
         self.positional_encodings = nn.Parameter(torch.randn(5000,d_model))
         
         self.transformer = nn.Transformer(
             d_model=d_model,
             nhead=nhead,
             num_decoder_layers=num_layers,
             num_encoder_layers=num_layers,
             batch_first=True
         )   
         self.fully_connected = nn.Linear(d_model,vocab_size)
    
    def forward(self, source, target):
        
        source = self.embedding(source)
        target = self.embedding(target)
        
        source = source + self.positional_encodings[:source.size(1)]
        target = target + self.positional_encodings[:target.size(1)]
        
        tgt_seq_len = target.size(1)
        tgt_mask = self.transformer.generate_square_subsequent_mask(tgt_seq_len).to(target.device)
        
        # Pass tgt_mask to the transformer
        output = self.transformer(source, target, tgt_mask=tgt_mask)

        output = self.transformer(source, target)
        return self.fully_connected(output)
                   
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)
```

```python
model = Custom_Transformer(vocab_size=28000, d_model=256, nhead=4, num_layers=6).to(device)


loss_fun = nn.CrossEntropyLoss(ignore_index=0)
optimizer = torch.optim.Adam(model.parameters(),lr =3e-4)
```

```python
def evaluate(model, val_dataLoader, loss_fun, device):
    model.eval()
    total_loss = 0

    with torch.no_grad():
        for source, target in val_dataLoader:
            source = source.to(device)
            target = target.to(device)

            output = model(source, target[:, :-1])

            output = output.reshape(-1, output.shape[-1])
            target = target[:, 1:].reshape(-1)

            loss = loss_fun(output, target)
            total_loss += loss.item()

    return total_loss / len(val_dataLoader)
```

```python
def translate_sentence(sentence, model, sp, device, max_len=50):

    model.eval()

    tokens = sp.encode(sentence, out_type=int)

    src = torch.tensor([tokens]).to(device)

    tgt_tokens = [sp.bos_id()]

    for _ in range(max_len):

        tgt_tensor = torch.tensor([tgt_tokens]).to(device)

        with torch.no_grad():
            output = model(src, tgt_tensor)

        next_token = output[0, -1].argmax().item()

        tgt_tokens.append(next_token)

        if next_token == sp.eos_id():
            break

    translation = sp.decode(tgt_tokens[1:-1])

    return translation
```
